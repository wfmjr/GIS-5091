### Import modules for problem

```python
import arcpy, os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from arcgis.features import SpatialDataFrame
from arcgis.gis import GIS

gis = GIS("http://slustl.maps.arcgis.com/home", "wfmjr68")
```

### Retrieving the Tapestry Feature Service

```python
tap_lyr = gis.content.get('9a57846e763c4bec87f96463e36f964c')
tap_lyr.layers
```

### Loading Tapestry Feature Service for Missouri Counties only into a spatial dataframe

```python
fl = tap_lyr.layers[1]

df = fl.query(where="STATE_NAME='Missouri'").df
df.head()
```

### 1. For the counties in Missouri only, what is the dominant Tapestry LifeMode Group Name(TLIFENAME)?

Here I used the mode function which retrieves the most repeated value of the TLIFENAME column.

```python
tlife_df = pd.DataFrame(df)
tlife_df.TLIFENAME.mode()
```

### 2. Create a histrogram that shows the number of Missouri counties per Tapestry LifeMode Group Name(TLIFENAME)?

Here I used groupby to aggregate the values in the TLIFENAME column and created a 'count' column to add the aggregation. The values in the 'count' column will be integers. After creating the 'count' column I was able to create barplot depicting aggregation.

```python
gb = tlife_df.groupby(['TLIFENAME'])
tlife_df['count'] = gb['TLIFENAME'].transform('count')
tlife_df.drop_duplicates(['TLIFENAME'],keep='first',inplace=True)
tlife_df.sort_values(['count'],ascending=False,inplace=True)
tlife_df.reset_index()

ax = sns.barplot(x='TLIFENAME', y='count', data=tlife_df, alpha=.5)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
ax.plot()
```

### 3. Create a pie chart that shows the number of Missouri counties per Tapestry LifeMode Group Name(TLIFENAME)?

I had a donut pie chart that looked similar, but not as clear. I found code that looked nice and modified it to work off of my data.

```python
fig, ax = plt.subplots(figsize=(12, 8), subplot_kw=dict(aspect="equal"))

tlifegrp = []
for name in tlife_df['TLIFENAME']:
    tlifegrp.append(name)
    
data = []   
for item in tlife_df['count']:
    data.append(item)

explode = (0.01,) * 8
wedges, texts=ax.pie(data, explode=explode, wedgeprops=dict(width=0.5), startangle=-40)

bbox_props = dict(boxstyle="square,pad=0.1", fc="w", ec="k", lw=0.72)
kw = dict(xycoords='data', textcoords='data', arrowprops=dict(arrowstyle="-"),
          bbox=bbox_props, zorder=0, va="center")

for i, p in enumerate(wedges):
    ang = (p.theta2 - p.theta1)/2. + p.theta1
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle = "angle,angleA=0,angleB={}".format(ang)
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    ax.annotate(tlifegrp[i], xy=(x, y), xytext=(1.35*np.sign(x), 1.4*y),horizontalalignment=horizontalalignment, **kw)

ax.set_title("Tapestry Life Name Groups")

plt.show()
```

### 4. For the counties in Missouri, what is the dominant Dominant Tapestry Segment Name (TSEGNAME)?

For the next three questions see explanation for the previous three questions.

```python
tseg_df = pd.DataFrame(df)
tseg_df.TSEGNAME.mode()
```

### 5. Create a histogram that shows the number of Missouri counties per Tapestry Segment Name (TSEGNAME)?

```python
gb = tseg_df.groupby(['TSEGNAME'])
tseg_df['count'] = gb['TSEGNAME'].transform('count')
tseg_df.drop_duplicates(['TSEGNAME'],keep='first',inplace=True)
tseg_df.sort_values(['count'],ascending=False,inplace=True)
tseg_df.reset_index()

ax = sns.barplot(x='TSEGNAME', y='count', data=tseg_df, alpha=.5)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
ax.plot()
```

### 5.Create a pie chart that shows the number of Missouri counties per Tapestry Segment Name (TSEGNAME)?

```python
fig, ax = plt.subplots(figsize=(12,10), subplot_kw=dict(aspect="equal"))

tseggrp = []
for name in tseg_df['TSEGNAME']:
    tseggrp.append(name)
    
data = []   
for item in tseg_df['count']:
    data.append(item)

wedges, texts = ax.pie(data, wedgeprops=dict(width=0.5), startangle=15)

bbox_props = dict(boxstyle="square,pad=0.1", fc="w", ec="k", lw=0.72)
kw = dict(xycoords='data', textcoords='data', arrowprops=dict(arrowstyle="-"),
          bbox=bbox_props, zorder=0, va="center")

for i, p in enumerate(wedges):
    ang = (p.theta2 - p.theta1)/2. + p.theta1
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle = "angle,angleA=0,angleB={}".format(ang)
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    ax.annotate(tseggrp[i], xy=(x, y), xytext=(1.35*np.sign(x), 1.4*y),horizontalalignment=horizontalalignment, **kw)

ax.set_title("Tapestry Segment Names")

plt.show()
```

### 7. Is there Any correlation between LifeMode Group and Segment name?

Here I used pandas crosstab function to show the number of Segment names that correspond to each LifeMode group per county. This also shows the total of each LifeMode group to each Segment name.

```python
grp_seg = pd.crosstab([df.NAME], [df.TLIFENAME, df.TSEGNAME], margins=True, margins_name='Total')
grp_seg
```

This a plot depicting the dataframe table shown above.

```python
sns.heatmap(pd.crosstab([df.NAME], [df.TLIFENAME, df.TSEGNAME]),
            cmap="YlGnBu", annot=True, cbar=False)
```
