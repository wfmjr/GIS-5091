### Importing modules

```python
import arcgis, arcpy, os, shapefile
import pandas as pd
from arcgis.gis import GIS
from arcgis.features import SpatialDataFrame
from arcgis.geoenrichment import *

gis = GIS("http://slustl.maps.arcgis.com/home", "wfmjr68")
```

### Taking the csv file and turning it into a Pandas dataframe for better data manipulation. Also sorted it descending by magnitude to show the most severe earthquakes first.

```python
df = pd.read_csv('significant_month.csv')
df_eqs = df.sort_values('mag',ascending=False)
df_reset = df_eqs.reset_index()
df_reset
```

### Making a list of coordinates from the latitude and longitude fields.

```python
eqsList = df_reset.iloc[:,2:4].values.tolist()
eqsList
```

### Taking the dataframe and making a spatial dataframe with wgs84 spatial reference. Also sorting by magnitude and reseting the index to later pull the top 3 earthquakes by magnitude.

```python
wkid = 4326
sdf = SpatialDataFrame(df_reset,geometry=[arcgis.geometry.Geometry({'x':r[1], 'y':r[0], 
                        'spatialReference':{'wkid':wkid}}) for r in eqsList])
sdf_eqs = sdf.sort_values('mag',ascending=False)
sdf_eqs.head()
```

### Taking the top 3 earthquakes only and placing them in a spatial dataframe with a SHAPE field.

```python
sigeqs_top3 = sdf_eqs[0:3]

sdf = sigeqs_top3[sigeqs_top3.SHAPE.notna()].copy()
sdf
```

### Creating a point feature layer from the top 3 earthquakes spatial dataframe.

```python
path = arcpy.env.workspace = 'D:/GIS-5091/Project 3/Project 3.gdb'
top3_eqs = 'top3_earthquakes'

sdf.to_featureclass(path,top3_eqs)
```

### Creating a multi ring buffer from the top 3 earthquakes feature layer.

```python
mr_buff = 'D:/GIS-5091/Project 3/mmi_buffers3.shp'
mmi_buffers = arcpy.MultipleRingBuffer_analysis(top3_eqs, mr_buff, [5,10,20,30,50,110,180,],'Miles', '', 'ALL')
```

### Loading the multi ring buffer into a spatial dataframe to plot to the map. 

```python
sdf = SpatialDataFrame.from_featureclass(mr_buff)
sdf
```

### Adding a new field showing the Modified Mercalli Intensity scale (MMI) of an earthquake.

```python
mmi_list = ['extreme','violent','severe','strong','moderate','light','weak']
sdf['MMI']=mmi_list
sdf
```
