```python
df
```

```python
import arcgis
from arcgis.gis import GIS
from IPython.display import display
```

```python
gis = GIS("http://slustl.maps.arcgis.com/home", "gregbrunner_slugis")
```

```python
resettlment_stats = gis.content.import_data(df, {"Address" : "country_of_origin_en"})
```

```python
map  =  gis.map('England', zoomlevel=7)
map
```

```python
map.add_layer(resettlment_stats, { "renderer":"ClassedSizeRenderer", "field_name":"value" })
```

```python
resettlment_stats.query().df
```

```python
sdf = resettlment_stats.query().df
```

```python
reset_map = gis.content.import_data(sdf, title='UNHCR Resettlements in France')
```

```python
new_map = gis.map()
new_map
```

```python
new_map.add_layer(reset_map, { "renderer":"ClassedSizeRenderer", "field_name":"value" })
```
