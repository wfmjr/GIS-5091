```python
aus_df = pd.read_json('http://popdata.unhcr.org/api/stats/asylum_seekers_monthly.json?year=2015,2016&month=8&coo=SYR')
```

```python
aus_df
```

```python
import arcgis
from arcgis.gis import GIS
from IPython.display import display
```

```python
gis = GIS('http://slustl.maps.arcgis.com/home', 'wfmjr68')
```

```python
resettlment_stats = gis.content.import_data(aus_df, {"Address" : "country_of_origin_en"})
```

```python
map = gis.map('AUS', zoomlevel = 4)
map
```

```python
map.add_layer(resettlment_stats, { "renderer":"ClassedSizeRenderer", "field_name":"value" })
```
