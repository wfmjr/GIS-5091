```python
from arcgis.gis import GIS
from arcgis.features.analysis import *
gis = GIS('https://slustl.maps.arcgis.com/home', 'wfmjr68')
```

```python
crime_search = gis.content.search("St. Louis Crime Data WFM", "feature layer", max_items=3)
crime_item = crime_search[1]
crime_item
```

```python
map1 = gis.map("St. Louis")
map1
```

```python
map1.add_layer(crime_item)
```

```python
analysis.calculate_density(crime_item, field=None, cell_size=None, cell_size_units='Meters', radius=None, radius_units=None, bounding_polygon_layer=None, area_units=None, classification_type='EqualInterval', num_classes=10, output_name='St_Louis_Crime_Density_JN')
```
