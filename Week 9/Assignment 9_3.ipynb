```python
import pandas as pd
import requests
```

```python
url = 'http://popdata.unhcr.org/api/stats/asylum_seekers_monthly.json?year=2015&month=1&coa=FRA'
data = requests.get(url)
as_json = data.json()
df = pd.DataFrame(as_json)
df.head()
```

### Question 1:
#### From what country did France receive the most immigrants in January 2015?

```python
max_ctry = df[df['value']==df['value'].max()]
print(max_ctry.country_of_origin_en.to_string(index=False))
```

### Question 2
#### What were the 5 countries that contributed the most immigrants in January of 2015?

```python
columns_1 = df[['country_of_origin_en','value']]
q2 = columns_1.sort_values(by='value', ascending=False)
q2.head()
```

### Question 3
#### From what countries did only 1 immigrant arrive from in January 2015?

```python
df.groupby('country_of_origin_en').filter(lambda x:x['value'] == 1)
```
