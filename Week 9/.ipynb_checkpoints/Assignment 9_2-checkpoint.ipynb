```python
import pandas as pd

with open('all_week_end_nov_4.csv', 'r') as file:
    last_week = pd.read_csv(file)
with open('all_week_aug_13_20.csv', 'r') as file2:
    week_old = pd.read_csv(file2)

week_old.head()
```

```python
df1 = pd.DataFrame(last_week)
df2 = pd.DataFrame(week_old)

comb_df = df1.append(ignore_index=True, other=df2)

locations = comb_df.iloc[:,1:3].values.tolist()
locations
```

```python
import arcgis

from arcgis.features import SpatialDataFrame

sdf = SpatialDataFrame(comb_df.set_geometry([arcgis.geometry.Geometry({'x':r[1], 'y':r[0],'spatialReference':{'wkid':4326}}) for r in locations]))
sdf.reset_index(inplace=True, drop=True)
sdf.to_featureclass(out_location='D:/GIS 5091/Week 9/week_9.gdb', out_name='EQs')
```
