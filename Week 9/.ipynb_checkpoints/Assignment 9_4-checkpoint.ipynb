```python
import pandas as pd
import requests
```

```python
def unhcr(year, month, country):
    global df
    url = ('http://popdata.unhcr.org/api/stats/asylum_seekers_monthly.json?year={}&month={}&coa={}').format(year,month,country)
    data = requests.get(url)
    as_json = data.json()
    df = pd.DataFrame(as_json)
    df.head()
unhcr(year= input('Enter year:'),month= input('Enter month by number:'),country= input('Entry country by ISO Alpha-3 code in all caps:'))
df.head()
```

### Country where the most immigrants came from.

```python
max_ctry = df[df['value']==df['value'].max()]
print(max_ctry.country_of_origin_en.to_string(index=False))
```

### The 5 countries that contributed the most immigrants.

```python
columns = df[['country_of_origin_en','value']]
q2 = columns.sort_values(by='value', ascending=False)
q2.head()
```

### Countries that had only 1 immigrant arrive.

```python
df.groupby('country_of_origin_en').filter(lambda x:x['value'] == 1)
```
