```python
import pandas as pd

with open('all_day.csv', 'r') as file:
    earthquakes = pd.read_csv(file)
earthquakes.head()
```

```python
df = pd.DataFrame(earthquakes)
eq_df = df.drop(['nst','gap','dmin','rms','net'], axis=1)
eq_df.head()
```

```python
locations = eq_df.iloc[:,1:3].values.tolist()
locations
```

```python
import arcgis
from arcgis.features import SpatialDataFrame

sdf = SpatialDataFrame(eq_df.set_geometry([arcgis.geometry.Geometry({'x':r[1], 'y':r[0],'spatialReference':{'wkid':4326}}) for r in locations]))
sdf1 = sdf[sdf['mag']<2.5]
sdf2 = sdf[(sdf['mag']>=2.5) & (sdf['mag']<4)]
sdf3 = sdf[(sdf['mag']>=4) & (sdf['mag']<5)]
sdf4 = sdf[sdf['mag']>=5]
sdf4
```

```python
sdf4.reset_index(inplace=True,drop=True) 
```

```python
from arcgis.gis import GIS
from arcgis.mapping import WebMap
from arcgis.features import analysis
gis = GIS('https://slustl.maps.arcgis.com/home', 'wfmjr68')
```

```python
eq = gis.content.import_data(sdf4, title='Magnitude Greater than 5')
eq.share(org = True)
```

```python
eq_webmap = WebMap()

eq_webmap.add_layer(eq, options={'title':'Magnitude Greater than 5'})

webmap_item_properties = {'title':'Magnitude Greater than 5',
                         'snippet':'Map created using USGS earthquake data from the past day.',
                         'tags':['earthquake', 'USGS', 'magnitude', 'csv']}

eq_webmap.save(webmap_item_properties, thumbnail='https://www.pitandquarry.com/wp-content/uploads/2016/05/usgs-logo-green.jpg')
```
