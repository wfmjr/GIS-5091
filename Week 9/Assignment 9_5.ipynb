```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('fsi-2018.xlsx')
df.head()
```

```python
hdr_list = list(df.columns[4:])
hdr_list
```

```python
list_of_lists = []
for column in df.columns[4:]:
    min_sort = df.sort_values(column)
    min_ten = min_sort.head(10)
    min_val = min_ten['Country'].values.tolist()
    list_of_lists.append(min_val)
print(list_of_lists)
```

```python
zip_final = zip(hdr_list, list_of_lists)
final= dict(zip_final)
final
```
